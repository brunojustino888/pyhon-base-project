Importando csv de uma determinada URL:

```
import pandas as pd
url = "https://raw.githubusercontent.com/brunojustino888/pyhon-base-project/main/dados_imoveis.csv";
dados = pd.read_csv(url);
```

```
print(dados.sample(10));
```

Amostra das informacoes do DF (Data Frame):

```
print(dados.info());
```

```
print(dados.Metragem.mean());
```

Obtem a quantidade de imoveis do bairro Vila Mariana no DF:

Obtem um DF do bairro Vila Mariana a partir do DF de imoveis:

```
df_vilaMariana = (dados.Bairro=="Vila Mariana");
print(dados[df_vilaMariana]);
```

Obtem a media da metragem dos imoveis do bairro Vila Mariana a partir do DF de imoveis:

```
dadosVM = dados[df_vilaMariana];
print(dadosVM.Metragem.mean());
```

Obter resumos dos bairros no DF:

```
print(dados['Bairro'].value_counts());
```

Grafico em barras de uma amostra de 10 bairros e metragem de  imoveis:

```
pandasDados = dados['Bairro'].value_counts();
print(pandasDados.head(10).plot.bar());
```

Obtem a media metragem para cada um dos bairros:

```
bairro_mediaMetragem = dados.groupby('Bairro')['Metragem'].mean();
print(bairro_mediaMetragem);
```

Formas de selecionar os dados por coluna especifica:

```
print(dados.loc[dados["Bairro"]=="Água Fria"]);
print(dados[dados["Bairro"] == "Água Branca"]);
print(dados[dados["Bairro"].isin(["Água Branca","Água Fria"])]);
```

Obtem a media de quartos por bairro e mostra no grafico que os bairros Vila Barbosa e Parada Inglesa sao aonde se encontram casas com maiores quantidade de comodos:

```
mediaQuartosPorbairro = dados.loc[:,["Bairro", "Quartos"]].groupby("Bairro").mean().sort_values("Quartos", ascending = True)
print(mediaQuartosPorbairro.head(8).plot.barh());
```

Obtem a media de vagas por bairro e mostra no grafico que os bairros se encontram com a quantidade de vagas iguais:

```
mediaVagasPorbairro = dados.loc[:,["Vagas","Bairro"]].groupby("Bairro").mean().sort_values("Vagas",ascending = True)
print(mediaVagasPorbairro.head(8).plot.pie(subplots=True, figsize=(11, 6)));
```

Obtem outras estatísticas dos dados, média, mediana, min, max:

```
print(dados[['Quartos', 'Banheiros']].describe().T);
print(dados[['Vagas']].median());
```
